```python
import requests

url = "https://data4library.kr/api/loanItemSrch" # 뒤에 정보 입력
r = requests.get(url) #2831d5955f6444f7448138766e4f76377b18e9ca7bbb14d265494ec4d18c7ef5

data = r.json() #data변수에 json형태로 저장
```

```python
books = []

for d in data['response']['docs']:
  books.append(d['doc'])
```

```python
import pandas as pd

books_df = pd.DataFrame(books) #books 배열을 데이터프레임으로 변환

books_df # 출력
```

```python
books = books_df.loc[:,'no':'isbn13'] # no ~ isbn13까지만 남기고 필요없는 데이터는 가져오지않는다.
books.head() # 가져올 정보의 갯수
```
